### LANGKAH 2: Menentukan Fungsi Keanggotaan (Membership Function)

```python
import skfuzzy as fuz
import matplotlib.pyplot as plt
```

```python
# Generate Triangular Membership Function
# skfuzzy.membership.trim(x, abc)
```

```python
def OurRange (_lo,_hi,_step):
    ourr = np.arange(_lo,_hi,_step)
    return ourr
```

```python
def MembershipFunc1 (_rule, _our_range, _title, _label0, _label1, _label2):
    low = fuz.trimf(_our_range, _rule[0])
    middle = fuz.trimf(_our_range, _rule[1])
    high = fuz.trimf(_our_range, _rule[2])
    
    fig, ax = plt.subplots(nrows=1, figsize=(6,3))
    ax.plot(_our_range, low, 'b', linewidth= 1.5, label=_label0)
    ax.plot(_our_range, middle, 'g', linewidth= 1.5, label=_label1)
    ax.plot(_our_range, high, 'r', linewidth= 1.5, label=_label2)
    
    ax.set_title(_title)
    ax.legend()
    
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.get_xaxis().tick_bottom ()
    ax.get_yaxis().tick_left()
    
    plt.tight_layout()
    plt.show()
    
    return low, middle, high

def MembershipFunc2 (_rule, _our_range, _title, _label0, _label1, _label2):
    low = fuz.trapmf(_our_range, _rule[0])
    middle = fuz.trimf(_our_range, _rule[1])
    high = fuz.trapmf(_our_range, _rule[2])
    
    fig, ax = plt.subplots(nrows=1, figsize=(6,3))
    ax.plot(_our_range, low, 'b', linewidth= 1.5, label=_label0)
    ax.plot(_our_range, middle, 'g', linewidth= 1.5, label=_label1)
    ax.plot(_our_range, high, 'r', linewidth= 1.5, label=_label2)
    
    ax.set_title(_title)
    ax.legend()
    
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.get_xaxis().tick_bottom ()
    ax.get_yaxis().tick_left()
    
    plt.tight_layout()
    plt.show()
    
    return low, middle, high
```

```python
xipk = OurRange(0,4,0.01)
ripk = np.array([
    [0,0,1.4,2.8],
    [2.4,2.9,3.5],
    [3.2,3.8,4,4]
], dtype = object)

lo_ipk, mi_ipk, hi_ipk = MembershipFunc2(ripk, xipk, 'IPK','Rendah','Sedang','Tinggi')
```

```python
xips = OurRange(0,4,0.01)
rips = np.array([
    [0,0,1.4,2.8],
    [2.4,2.9,3.5],
    [3.2,3.8,4,4]
], dtype = object)



lo_ips, mi_ips, hi_ips = MembershipFunc2(rips, xips, 'IPS','Rendah','Sedang','Tinggi')
```

```python
xto = OurRange(1,9,1)
rto = np.array([
    [1,1,3],
    [2,4,5],
    [4,8,8]
])

lo_to, mi_to, hi_to = MembershipFunc1(rto, xto, 'Tanggungan Orang Tua','Sedikit','Sedang','Banyak')
```

```python
xukt = OurRange(500000,15000000,500000)
rukt = np.array([
    [500000,500000,2000000,4000000],
    [2000000,5000000,8000000],
    [6000000,10000000,15000000,15000000]
], dtype=object)

lo_ukt, mi_ukt, hi_ukt = MembershipFunc2(rukt, xukt, 'Biaya UKT','Rendah','Sedang','Tinggi')
```

### LANGKAH 4: Menentukan Status Keanggotaan

```python
def MembershipDeg1 (_range, _lo, _mi, _hi, _val):
    low = fuz.interp_membership(_range, _lo, _val)
    middle = fuz.interp_membership(_range, _mi, _val)
    high = fuz.interp_membership(_range, _hi, _val)
    
    return low, middle, high 
```

```python
def Status1 (_membership,_label0,_label1,_label2):
    status = ''
    if _membership[0]> _membership[1] and _membership[0] > _membership[2]:
        status = _label0
    elif _membership[1]> _membership[0] and _membership[1] > _membership[2]:
        status = _label1
    elif _membership[2]> _membership[0] and _membership[2] > _membership[1]:
        status = _label2
    return status
```

```python
ipk_member = []

for i in range(len(ipk)):
    ipk_member.append(MembershipDeg1(xipk, lo_ipk, mi_ipk, hi_ipk, ipk[i]))
    print(f"ipk_member_{i+1}: {ipk_member[i]}")

ipk_status = []
for i in range(len(ipk)):
    ipk_status.append(Status1(ipk_member[i],'rendah', 'sedang','tinggi'))
    print(f"ipk_status_{i+1}: {ipk_status[i]}")
```

```python
ips_member = []

for i in range(len(ips)):
    ips_member.append(MembershipDeg1(xipk, lo_ips, mi_ips, hi_ips, ips[i]))
    print(f"ips_member_{i+1}: {ips_member[i]}")

ips_status = []
for i in range(len(ipk)):
    ips_status.append(Status1(ips_member[i],'rendah', 'sedang','tinggi'))
    print(f"ips_status_{i+1}: {ips_status[i]}")
```

### LANGKAH 5: Membuat baris pengetahuan fuzzy (Fuzzy Rule-Based)

```python
def RuleBased(_ipk_status, _ips_status):
    beasiswa_status = ""
    if ((_ips_status == 'rendah' or _ips_status == 'sedang') and _ipk_status == 'rendah'):
        beasiswa_status = "tidak diterima"
    elif ((_ips_status == 'sedang' or _ips_status == 'tinggi') and (_ipk_status == 'tinggi' or _ipk_status == 'sedang')):
        beasiswa_status = "diterima"
    else:
        beasiswa_status = "unknown"
    return beasiswa_status
```
